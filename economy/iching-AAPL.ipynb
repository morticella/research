```python


def market(row):
    if row['delta'] > 0:
        return 1
    return 0
df['up'] = df.apply(market, axis=1)
def valueToScore(val):
    borders = [15, 25, 30, 70, 75, 85]
    scores = [-3,-2,-1,0,1,2,3]
    for border, score in zip(borders, scores):
        if val <= border:
            return score
    return scores[-1]

def computeStatistic(row):
    rate = row['tempRate']
    currentStatistic = row['ingeneo']
    if currentStatistic is None:
        currentStatistic = 0
    newStatistic = valueToScore(rate)
    if currentStatistic > 0:
        if newStatistic <= 0:
            return currentStatistic
        elif currentStatistic <= newStatistic: 
            return newStatistic
        else:
            return currentStatistic
    elif currentStatistic < 0:
        if currentStatistic != -4:
            if newStatistic != 0:
                return newStatistic
            else:
                return currentStatistic
        else:
            return currentStatistic
    elif currentStatistic == 0: 
        return newStatistic

def computeBuggyPercentageFactory(df, field1, field2):
    numPRs = df.pivot_table(index=[field1, field2,"up"], aggfunc='size');
    def computeBuggyPercentage(row):
        buggy = notBuggy = 0
        try:
            buggy = numPRs[ row[field1], row[field2], 1] 
        except KeyError:
            pass
        try:
            notBuggy = numPRs[ row[field1], row[field2], 0] 
        except KeyError:
            pass
        total = buggy + notBuggy

        if buggy != 0 and total > 1:
            if (100 * buggy / total) >= 70 or (100 * buggy / total) <= 30:
                row['prediction_reason_rate'].append({'rate': 100 * buggy / total, 'feature': [field1, field2]});
            return 100 * buggy / total
        elif buggy == 0 and total > 1:
            row['prediction_reason_rate'].append({'rate': 0, 'feature': [field1, field2]});
            return 0
        else:
            return 50
        # otherwise it's None

    return computeBuggyPercentage

def computeBuggyPercentageFactoryTrilogy(df, field1, field2, field3):
    numPRs = df.pivot_table(index=[field1, field2, field3,"up"], aggfunc='size');
    def computeBuggyPercentage(row):
        buggy = notBuggy = 0
        try:
            buggy = numPRs[ row[field1], row[field2], row[field3], 1] 
        except KeyError:
            pass
        try:
            notBuggy = numPRs[ row[field1], row[field2], row[field3], 0] 
        except KeyError:
            pass
        total = buggy + notBuggy
        if buggy != 0 and total > 1:
            
            if (100 * buggy / total) >= 70 or (100 * buggy / total) <= 30:
                row['prediction_reason_rate'].append({'rate': 100 * buggy / total, 'feature': [field1, field2, field3]});
            return 100 * buggy / total
        elif buggy == 0 and total > 1:
            row['prediction_reason_rate'].append({'rate': 0, 'feature': [field1, field2, field3]});
            return 0
        else:
            return 50
        # otherwise it's None

    return computeBuggyPercentage
```

```python
df['ingeneo'] = 0
data = [[] for i in range(len(df))]
df['prediction_reason_rate'] = data

two_features = [
    {'feat_a': 'exa','feat_b': 'up_prev'},
    {'feat_a': 'exa','feat_b': 'up_next'},
    {'feat_a': 'up_next','feat_b': 'up_prev'}
]

for feature in two_features:
    computeBuggyPercentage = computeBuggyPercentageFactory(df, feature['feat_a'], feature['feat_b'])
    df['tempRate'] = df.apply(computeBuggyPercentage, axis=1)
    df['ingeneo'] = df.apply(computeStatistic, axis=1)

three_features = [
    {'feat_a': 'exa','feat_b': 'up_prev', 'feat_c':'up_next'},
]
for feature in three_features:
    computeBuggyPercentage = computeBuggyPercentageFactoryTrilogy(df, feature['feat_a'], feature['feat_b'], feature['feat_c'])
    df['tempRate'] = df.apply(computeBuggyPercentage, axis=1)
    df['ingeneo'] = df.apply(computeStatistic, axis=1)
df[0:100]
```
